Почему прямые не перпендикулярны

```python
x = np.linspace(-10,10,2)
y = 3*x+1
y2=(-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.axis((-10,10,-10,10))
plt.gca().set_aspect('equal')
```

Код для построения графиков окружности, элипса, гиперболы.

```python
r = 6
x = np.linspace(-r,r,1000)
y = np.sqrt(-x**2+r**2)
plt.plot(x, y)
plt.plot(x,-y)
plt.gca().set_aspect('equal')
```

```python
a=4
b=2
x = np.linspace(-10, 10, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
plt.contour(x, y,(x**2/a**2 + y**2/b**2),[1])
```

Трехмерный график двух параллельных плоскостей
Трехмерный график двух любых поверхностей второго порядка

```python
from pylab import *
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z = 2*X + 5*Y + 5
Z1 = 2*X + 5*Y +20
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z1, color = 'k')
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z = np.sqrt(X**2, Y**2)
Z1= np.sqrt(X**2 + Y**2,X**2 - Y**2 )
ax.plot_surface(X,Y,Z)
ax.plot_surface(X,Y,Z1)
```
